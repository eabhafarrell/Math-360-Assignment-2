We attempted different combinations of $a$, $b$, $c$ values while setting January's average daily customer, average temperature, humidity, and precipitation to be our reference points. And we picked the one combination that gives us the closest value to July average daily customer, which is $a = -3, b = 32,$ and $ c = -1$.

We then separate the cases into the two different stores because the average price for different store is different. And we are able to sum them together later because the expected value of the sum is the sum of the expected values.

**Alumni**

```python
N = 5000
C0 = jan_daily_mean_alu
T0 = mu_jan_tem
P0 = mu_jan_precip
H0 = mu_jan_humid
a  = -3
b = 32
c = -1
T = np.random.normal(loc = mu_july_tem, scale = sig_july_tem, size = N)
P = np.random.exponential(scale = mu_july_precip, size = N)
H = np.random.normal(loc = mu_july_humid, scale = sig_july_humid, size = N)
num = C0 + a * (T - T0)**2 - b * (P - P0) - c * (H - H0)**2
C1 = np.maximum(num, 0)
print("a = ", a)
print("b = ", b)
print("c = ", c)
print("mean = ", C1.mean())
            

kde_Y = stats.gaussian_kde(C1)

plt.hist(C1,bins=50,density=True)

t = np.linspace(0,1700,3000)
plt.plot(t,kde_Y(t))

plt.show()
```

We then use what we have for a, b, c for other months.

```python
#we define a function so that we don't have to repeat the thing again and again
def monte_carlo_monthly_mean(mu_tem, sd_tem, mu_precip, mu_humid, sd_humid,
                   C0, T0 = mu_jan_tem, P0 = mu_jan_precip, H0 = mu_jan_humid, a = -3, b = 32, c = -1,
                   N=2000):
    T = np.random.normal(loc=mu_tem, scale=sd_tem, size=N)
    P = np.random.exponential(scale=mu_precip, size=N)
    H = np.random.normal(loc=mu_humid, scale=sd_humid, size=N)
    num = C0 + a * (T - T0)**2 - b * (P - P0) - c * (H - H0)**2
    C1 = np.maximum(num, 0)
    return C1.mean()
```

```python
months = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]
month = np.arange(0,12,1)
month_days = [31, 28, 31, 30, 31, 30, 31,31,30,31,30,31]
mu_tem = [mu_jan_tem, mu_feb_tem, mu_march_tem, mu_april_tem,
        mu_may_tem, mu_june_tem, mu_july_tem, mu_august_tem,
        mu_sep_tem, mu_oct_tem, mu_nov_tem, mu_dec_tem]
sd_tem = [sig_jan_tem, sig_feb_tem, sig_march_tem, sig_april_tem,
        sig_may_tem, sig_june_tem, sig_july_tem, sig_august_tem,
        sig_sep_tem, sig_oct_tem, sig_nov_tem, sig_dec_tem]
mu_precip = [mu_jan_precip, mu_feb_precip, mu_march_precip, mu_april_precip,
        mu_may_precip, mu_june_precip, mu_july_precip, mu_august_precip,
        mu_sep_precip, mu_oct_precip, mu_nov_precip, mu_dec_precip]
sd_precip = [sig_jan_precip, sig_feb_precip, sig_march_precip, sig_april_precip,
        sig_may_precip, sig_june_precip, sig_july_precip, sig_august_precip,
        sig_sep_precip, sig_oct_precip, sig_nov_precip, sig_dec_precip]
mu_humid = [mu_jan_humid, mu_feb_humid, mu_march_humid, mu_april_humid,
        mu_may_humid, mu_june_humid, mu_july_humid, mu_august_humid,
        mu_sep_humid, mu_oct_humid, mu_nov_humid, mu_dec_humid]
sd_humid = [sig_jan_humid, sig_feb_humid, sig_march_humid, sig_april_humid,
        sig_may_humid, sig_june_humid, sig_july_humid, sig_august_humid,
        sig_sep_humid, sig_oct_humid, sig_nov_humid, sig_dec_humid]
```

Here are the mean customer frequency at Alumni Center for each month.

```python
#alumni one
monthly_mean_alu = []
for i in month:
    monthly_avg = monte_carlo_monthly_mean(mu_tem = mu_tem[i], sd_tem = sd_tem[i],
                               mu_precip = mu_precip[i],
                               mu_humid = mu_humid[i],sd_humid = sd_humid[i],
                               C0 = jan_daily_mean_alu)

    monthly_mean_alu.append(monthly_avg*month_days[i])
results_alu = pd.DataFrame({
    "Month": months,
    "Mean_customers": monthly_mean_alu
})
results_alu
```

Here are the mean customer frequency at Sauder for each month.

```python
#sauder
monthly_mean_sauder = []
for i in month:
    monthly_avg = monte_carlo_monthly_mean(mu_tem = mu_tem[i], sd_tem = sd_tem[i],
                               mu_precip = mu_precip[i],
                               mu_humid = mu_humid[i],sd_humid = sd_humid[i],
                               C0 = jan_daily_mean_sauder)
    monthly_mean_sauder.append(monthly_avg*month_days[i])


results_sauder = pd.DataFrame({
    "Month": months,
    "Mean_customers": monthly_mean_sauder
})
results_sauder
total_mean_customers_sauder = results_sauder["Mean_customers"].sum()
print(total_mean_customers_sauder)
```
